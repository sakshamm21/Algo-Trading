```python
def func(s,st,ed):
  D1=yf.download(s,start=st,end=ed,interval='1m')
  D1.reset_index(level=0,inplace=True)

  D1_high=D1[D1['High']==D1['High'].max()]['Datetime'].iloc[0]

  D1_low=D1[D1['Low']==D1['Low'].min()]['Datetime'].iloc[0]
    
  op=D1['Open'].iloc[0]

  k=0.35
  op=mth.floor((1-(k/100))*op)
  D1_op035=D1.iloc[abs(D1['Open']-op).argsort()[:1]]['Datetime'].iloc[0]

  k=0.5
  op=mth.floor((1-(k/100))*op)
  D1_op05=D1.iloc[abs(D1['Open']-op).argsort()[:1]]['Datetime'].iloc[0]

  k=1
  op=mth.floor((1-(k/100))*op)
  D1_op1=D1.iloc[abs(D1['Open']-op).argsort()[:1]]['Datetime'].iloc[0]

  k=2
  op=mth.floor((1-(k/100))*op)
  D1_op2=D1.iloc[abs(D1['Open']-op).argsort()[:1]]['Datetime'].iloc[0]

  k=3.5
  op=mth.floor((1-(k/100))*op)
  D1_op35=D1.iloc[abs(D1['Open']-op).argsort()[:1]]['Datetime'].iloc[0]

  df= {'Stock': s,'High':D1_high,'Low':D1_low,'P035':D1_op035, 'P05': D1_op05, 'P1': D1_op1, 'P2':D1_op2, 'P35':D1_op35}
  return df
```
